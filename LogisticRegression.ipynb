1. What is Logistic Regression, and how does it differ from Linear Regression ?

-> Logistic regression and linear regression are both statistical methods used for predictive modeling, but they differ in the type of dependent variable they handle.

-> Logistic regression is used for classification problems with categorical outcomes (like yes/no or 0/1), while linear regression is used for regression problems with continuous outcomes.

 ->Essentially, logistic regression predicts the probability of a categorical outcome, while linear regression predicts a continuous value.

2. What is the mathematical equation of Logistic Regression ?

-> The logistic regression equation is:

p = 1 / (1 + e^(-z))

where:

- p is the probability of the positive class (e.g., 1 or True)
- e is the base of the natural logarithm (approximately 2.718)
- z is the linear combination of the input features, given by:

-> z = β0 + β1_x1 + β2_x2 + … + βn*xn**

where:

- β0 is the intercept or bias term
- β1, β2, …, βn are the coefficients of the input features
- x1, x2, …, xn are the input features

-> This equation is also known as the sigmoid function or logit function, and it maps the input z to a probability between 0 and 1.

3. Why do we use the Sigmoid function in Logistic Regression ?

-> The sigmoid function is used in Logistic Regression to transform the output of the linear model into a probability between 0 and 1. This allows the model to predict the probability of a binary outcome (e.g., success or failure, yes or no).

4. What is the cost function of Logistic Regression ?
-> The cost function for logistic regression is typically the Log Loss or Cross-Entropy Loss, given by:

-> J(β) = -[y * log(p) + (1-y) * log(1-p)]

where:

- J(β) is the cost function
- y is the actual label (0 or 1)
- p is the predicted probability
- log is the natural logarithm

-> This cost function measures the difference between the predicted probabilities and the actual labels. The goal of logistic regression is to minimize this cost function using optimization algorithms like gradient descent.

5. What is Regularization in Logistic Regression? Why is it needed ?

-> Regularization is a technique used to prevent overfitting in logistic regression models by adding a penalty term to the cost function.

-> Types of Regularization:

-> There are two common types of regularization:

- L1 Regularization (Lasso): adds a penalty term proportional to the absolute value of the coefficients.
- L2 Regularization (Ridge): adds a penalty term proportional to the square of the coefficients.

-> Why Regularization is Needed:

- Prevent overfitting: by reducing the magnitude of the coefficients.
- Improve model generalization: by reducing the impact of noise in the data.
- Stabilize the model: by reducing the impact of correlated features.

-> By adding a penalty term, regularization encourages the model to find a balance between fitting the data well and keeping the coefficients small. This helps to improve the model's performance on unseen data.

6.  Explain the difference between Lasso, Ridge, and Elastic Net regression

-> Lasso, Ridge, and Elastic Net are regularization techniques used to prevent overfitting in regression models.

Key Differences
- Lasso Regression (L1 Regularization): adds a penalty term proportional to the absolute value of the coefficients. It can set coefficients to zero, effectively performing feature selection.
- Ridge Regression (L2 Regularization): adds a penalty term proportional to the square of the coefficients. It reduces the magnitude of coefficients but doesn't set them to zero.
- Elastic Net Regression: combines both L1 and L2 regularization. It can set coefficients to zero like Lasso and also reduces the magnitude of coefficients like Ridge.

7. When should we use Elastic Net instead of Lasso or Ridge ?

-> Elastic Net is a good choice when:

- You want to balance feature selection and coefficient shrinkage: Elastic Net combines the benefits of Lasso (feature selection) and Ridge (coefficient shrinkage).
-You have correlated features: Elastic Net can handle correlated features better than Lasso, which can arbitrarily select one feature over others.
- You want to avoid over-shrinkage: Elastic Net can reduce over-shrinkage of coefficients compared to Lasso.
- You're dealing with high-dimensional data: Elastic Net can handle high-dimensional data with a mix of relevant and irrelevant features.

-> When to Prefer Elastic Net over Lasso or Ridge
- Lasso is selecting too few features: Elastic Net can select more features than Lasso while still performing feature selection.
- Ridge is not providing enough regularization: Elastic Net can provide a stronger regularization effect than Ridge while still allowing for some feature selection.

8. What is the impact of the regularization parameter (λ) in Logistic Regression ?

-> The regularization parameter (λ) controls the strength of regularization in logistic regression.

-> Effects of λ
- Small λ: weak regularization, model fits training data well but may overfit.
- Large λ: strong regularization, model is simpler, but may underfit.

-> Impact on Model
- Coefficient magnitude: larger λ reduces coefficient magnitude.
- Model complexity: larger λ reduces model complexity.
- Bias-variance tradeoff: λ controls the tradeoff between bias (underfitting) and variance (overfitting).

-> Choosing λ
- The optimal value of λ is typically determined using cross-validation, where the value that results in the best performance on a validation set is chosen

9. What are the key assumptions of Logistic Regression ?

-> Logistic regression assumes:

- Independence of observations: each observation is independent of others.
- Linearity in the log-odds: the relationship between the log-odds of the outcome and the predictors is linear.
- No multicollinearity: predictors are not highly correlated with each other.
- No outliers: data does not contain extreme outliers that can affect model estimates.
- Binary outcome: the outcome variable is binary (0/1, yes/no, etc.).

-> Additional Considerations
- Sufficient sample size: the sample size is sufficient to estimate model parameters.
-  Correct model specification: the model includes all relevant predictors and interactions.

-> Violations of these assumptions can lead to biased or inefficient estimates, and may require alternative modeling approaches or data transformations.

10. What are some alternatives to Logistic Regression for classification tasks ?

-> Some popular alternatives to logistic regression for classification tasks include:

- Decision Trees: a tree-based model that splits data into subsets based on feature values.
- Random Forests: an ensemble model that combines multiple decision trees.
- Support Vector Machines (SVMs): a model that finds the hyperplane that maximally separates classes.

11. What are Classification Evaluation Metrics ?

-> Common metrics used to evaluate classification models include:

- Accuracy: proportion of correctly classified instances.
- Precision: proportion of true positives among all predicted positive instances.
- Recall: proportion of true positives among all actual positive instances.
- F1-score: harmonic mean of precision and recall.
- Area Under the ROC Curve (AUC-ROC): measures model's ability to distinguish between classes.
- Confusion Matrix: table summarizing predictions against actual outcomes.

12. How does class imbalance affect Logistic Regression ?

-> Class imbalance occurs when one class has a significantly larger number of instances than the other(s). This can affect logistic regression models in several ways:

-> Effects of Class Imbalance
- Biased predictions: the model may be biased towards the majority class.
- Poor minority class performance: the model may struggle to accurately predict the minority class.
- Thresholding issues: the default threshold of 0.5 may not be optimal for imbalanced data.

-> Consequences
- Reduced accuracy: for the minority class.
- Increased false negatives: for the minority class.

-> Addressing Class Imbalance
- Resampling techniques: oversampling the minority class, undersampling the majority class.
- Cost-sensitive learning: assigning different costs to misclassification errors.
- Thresholding adjustments: adjusting the decision threshold.
- Using class weights: assigning different weights to classes during training.

13. What is Hyperparameter Tuning in Logistic Regression ?

-> Hyperparameter tuning involves adjusting the model's hyperparameters to optimize its performance. In logistic regression, common hyperparameters to tune include:

- Regularization parameter (λ or C): controls the strength of regularization.
- Penalty type: choice between L1 or L2 regularization.

-> Hyperparameter Tuning Techniques
- Grid Search: exhaustively searching through a predefined set of hyperparameters.
- Random Search: randomly sampling hyperparameters from a defined distribution.
- Cross-Validation: evaluating model performance on unseen data to prevent overfitting.

-> Goals of Hyperparameter Tuning
- Improve model performance: optimize metrics such as accuracy, precision, or recall.
- Prevent overfitting: find a balance between model complexity and generalization.

14. What are different solvers in Logistic Regression? Which one should be used ?

-> Logistic regression solvers are algorithms used to optimize the model's parameters. Common solvers include:

- Liblinear: a coordinate descent algorithm, suitable for small to medium-sized datasets.
- LBFGS: a quasi-Newton algorithm, suitable for larger datasets.
- Newton-CG: a Newton method with conjugate gradient, suitable for large datasets.
- SAG: a stochastic average gradient algorithm, suitable for large datasets.
- SAGA: a variant of SAG with better convergence properties.

-> Choosing a Solver

-> The choice of solver depends on:

- Dataset size: larger datasets may require more efficient solvers like SAG or LBFGS.
- Computational resources: solvers like Liblinear may be faster on smaller datasets.
- Convergence speed: solvers like SAGA may converge faster than others

15. How is Logistic Regression extended for multiclass classification ?

-> Logistic regression can be extended for multiclass classification using:

- One-vs-Rest (OvR): train a separate logistic regression model for each class, where the model predicts the probability of a sample belonging to that class versus all other classes.
- One-vs-One (OvO): train a separate logistic regression model for each pair of classes, where the model predicts the probability of a sample belonging to one class versus the other.
- Multinomial Logistic Regression: extend the logistic regression model to predict probabilities for multiple classes simultaneously, using a softmax function to ensure probabilities sum to 1.

-> Key Considerations
- Class balancing: ensure that the classes are balanced or use techniques like class weighting.
- Model selection: choose the best approach based on the number of classes, dataset size, and performance metrics.


16. What are the advantages and disadvantages of Logistic Regression ?

- Feature correlation: can be affected by highly correlated features.
- Not suitable for high-dimensional data: may not perform well with a large number of features.

18. What is the difference between Softmax Regression and Logistic Regression ?

-> Softmax regression and logistic regression are both used for classification problems, but they differ in the number of classes they can handle:

-> Logistic Regression
- Binary classification: logistic regression is used for binary classification problems, where the target variable has two classes.
- Sigmoid function: logistic regression uses the sigmoid function to predict probabilities.

-> Softmax Regression
- Multiclass classification: softmax regression is used for multiclass classification problems, where the target variable has more than two classes.
- Softmax function: softmax regression uses the softmax function to predict probabilities for each class.

-> Key differences
- Number of classes: logistic regression is limited to binary classification, while softmax regression can handle multiclass classification.
- Activation function: logistic regression uses the sigmoid function, while softmax regression uses the softmax function.

-> When to use each
- Logistic regression: use for binary classification problems, such as spam vs. non-spam emails.
- Softmax regression: use for multiclass classification problems, such as handwritten digit recognition (0-9).

19. How do we choose between One-vs-Rest (OvR) and Softmax for multiclass classification ?

-> When deciding between OvR and Softmax for multiclass classification, consider the following factors:

-> One-vs-Rest (OvR)
- Advantages: simple to implement, can be used with any binary classifier.
- Disadvantages: may not perform well if classes are highly imbalanced or if the binary classifiers are not well-calibrated.

-> Softmax
- Advantages: can learn the relationships between all classes simultaneously, often performs better than OvR.
- Disadvantages: can be computationally expensive, may require larger datasets.

-> Choosing between OvR and Softmax
- Number of classes: Softmax may be more suitable for problems with a large number of classes.
- Class balance: OvR may struggle with highly imbalanced classes, while Softmax can handle class imbalance more robustly.
- Model complexity: Softmax can be more complex and require more data, while OvR can be simpler and more interpretable.
- Performance metrics: choose the approach that optimizes the desired performance metric (e.g., accuracy, F1-score).


20. How do we interpret coefficients in Logistic Regression?

-> In logistic regression, coefficients represent the change in the log-odds of the target variable for a one-unit change in the predictor variable, while holding all other variables constant.

-> Interpreting Coefficients
- Sign of coefficient: a positive coefficient indicates an increase in the log-odds, while a negative coefficient indicates a decrease.
- Magnitude of coefficient: the larger the coefficient, the stronger the relationship between the predictor and target variable.
- Odds ratio: exponentiating the coefficient gives the odds ratio, which represents the change in odds for a one-unit change in the predictor variable.

15. Write a Python program to apply feature scaling (Standardization) before training a Logistic Regression
model. Evaluate its accuracy and compare results with and without scaling
